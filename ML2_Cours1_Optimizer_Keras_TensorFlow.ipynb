```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
#on a fait units=1 puis 2
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[2])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs1 = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
xs2 = np.array([-2.0, 1.0, 0.0, 1.0, 4.0, 2.0], dtype=float)
#xs = tf.constant([xs1,xs2],tf.float16)
xs = np.column_stack((xs1, xs2))
ys = np.array([-9.0, 2.0, 1.0, 6.0, 17.0, 13.0], dtype=float)
```

```python
#on a fait epochs = 500 puis 1000
model.fit(xs,ys,epochs=500)
```

```python
print(model.predict([[10.0,-3.0]]))
```
