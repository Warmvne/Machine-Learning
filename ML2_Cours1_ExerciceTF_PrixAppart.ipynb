```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
#la fonction utilise un réseau de neurone shallow (une couche uniquement)
#le but du réseau est de déterminer le prix d'un logement selon le nombre de chambre qu'on choisira en paramètre
#le prix sans chambre est de 50K euros et chaque chambre en plus, le prix augmente de ce même montant.
def house_model(y_new):
  #nombre de chambre
  xs = [0,1,2,3,4,5,6]
  #prix selon le nbre de chambre en 10**2 K euros
  ys = [0.5,1.0,1.5,2.0,2.5,3.0,3.5]
  #création du réseau de neurones avec une seule couche et un paramètre d'entrée qui sera xs
  model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
  #compilation du modèle avec sgd comme optimisateur (on peut tester avec adam) et on calcule la loss avec la mse
  model.compile(optimizer='sgd', loss='mean_squared_error')
  #on entraîne le modèle
  model.fit(xs,ys,epochs=500)
  #renvoi la valeur prédite pour le paramètre choisi
  return model.predict(y_new)[0]
```

```python
#test pour prédire le prix d'un logemment avec 7 chambres
print(house_model([7.0]))
```

```python
#on obtient 4 en sortie ce qui signifie que le prix d'un logement de 7 pièces est de 400K euros, ce qui est logique
#d'après la définition de notre modèle
```
