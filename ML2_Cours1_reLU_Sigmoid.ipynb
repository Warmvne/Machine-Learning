```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from math import *
```

```python
import numpy as np
```

```python
def relu(x):
  return np.maximum(-x,x)
```

```python
def sigmoid(x):
  return 1/(1 + np.exp(-x))
```

```python
x_values = np.linspace(-5, 5, 100)
y_values = relu(x_values)
y_values1 = sigmoid(x_values)
```

```python
plt.plot(x_values, y_values, label='ReLU')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x_values, y_values1, label='Sigmoid')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.legend()
plt.grid(True)
plt.show()
```
